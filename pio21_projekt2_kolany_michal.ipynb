### Zadanie 1: password [30 pkt]

Zaprogramuj funkcję `password(arg=8)` z 1 domyślnym parametrem formalnym `arg=8`. W zależności od typu i wartości tego parametru funkcja powinna 

* gdy typ parametru `arg` to `int` oraz `arg >= 8` to funkcja generuje **poprawne hasło**
* gdy typ parametru `arg` to `str` to funkcja sprawdza, czy hasło jest **poprawne** (True/False)
* w innym przypadku funkcja zwraca `None`

**Poprawne hasło** zawiera
* minimum 8 znaków, w tym
    * minimum 1 literę dużą
    * minimum 1 literę małą
    * minimum 1 cyfrę
    * minimum 1 znak specjalny ze znaków `_+-!@#%^&*.,;[]()|`

Przykładowo
```
    >>> password()
    '*2R3y]Ft'
    >>> password(18)
    'GPvo3-;%Z;*F_|gp52'
    >>> password('*2R3y]Ft')
    True
    >>> password('GPvo3-';%Z;*F_|gp52')
    True
    >>> password('gG1+')
    False
    >>> password('AlaMaKota')
    False
```

````python
import string
import secrets

def checkPassword(password):
    try:
        return (any(c.isupper() for c in password)            # minimum 1 literę dużą
                and any(c.islower() for c in password)        # minimum 1 literę małą
                and any(c.isdigit() for c in password)        # minimum 1 cyfrę
                and any(not(c.isalnum()) for c in password)   # minimum 1 znak specjalny ze znaków _+-!@#%^&*.,;[]()|
                and len(password) >= 8)                       # minimum 8 znaków (sprawdzane na końcu, żeby powyższe rzeczy zdążyły wyrzucić type error w przypadku np. tablic o rozmiarze mniejszym niż 8, bo o dziwo kolejność ma tu znaczenie)
    except:
        return None
    

def password(arg=8):
    alphabet = string.ascii_letters + string.digits + '_+-!@#%^&*.,;[]()|'
    try:
        if arg<8:
            raise TypeError
        while True:
            password = ''.join(secrets.choice(alphabet) for i in range(arg))
            if (checkPassword(password)):
                break
        return password
    except TypeError:
        return checkPassword(arg)
        
print(password(11))
print(password(password(11)))
print(password([1,2,3]))
print(password(4))
````

### Zadanie 3: prawo Zipfa [30 pkt]

> **Prawo Zipfa** lub **prawo Estoupa-Zipfa** – prawo opisujące zasadę częstotliwości użycia w dowolnym języku poszczególnych wyrazów. Głosi ono, że jeżeli dla jakiegokolwiek tekstu lub grupy tekstów ustala się wykaz wyrazów ułożonych w malejącym porządku częstotliwości ich występowania, to ranga (numer porządkowy) wyrazu jest odwrotnie proporcjonalna do częstotliwości, zatem iloczyn częstotliwości i rangi powinien być wielkością stałą. (Wikipedia)

Zilustruj prawo Zipfa na

````python
import collections


def generateZipfTable(text, top):
    text = parseText(text)
    text = text.lower()
    top_word_frequencies = topWordFrequencies(text, top)

    return createZipfTable(top_word_frequencies)

def extractText(soup):
    lyrics = soup.find(id='songText')
    lyrics.h2.extract()
    lyrics.div.extract()
    lyrics.a.extract()
    return lyrics.getText()

def parseText(text): 
    chars_to_remove = "!\"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"
    tr = text.maketrans("", "", chars_to_remove)

    return (text.translate(tr))
    


def topWordFrequencies(text, top):
    words = text.split()
    word_frequencies = collections.Counter(words)
    top_word_frequencies = word_frequencies.most_common(top)

    return top_word_frequencies


def createZipfTable(frequencies):
    zipf_table = []
    top_frequency = frequencies[0][1]

    for index, item in enumerate(frequencies, start=1):
        relative_frequency = "1/{}".format(index)
        zipf_frequency = top_frequency * (1 / index)
        difference_actual = item[1] - zipf_frequency
        difference_percent = (item[1] / zipf_frequency) * 100

        zipf_table.append({"word": item[0],
                           "actual_frequency": item[1],
                           "relative_frequency": relative_frequency,
                           "zipf_frequency": zipf_frequency,
                           "difference_actual": difference_actual,
                           "difference_percent": difference_percent})

    return zipf_table


def printZipfTable(zipf_table):
    print("  Rank      Word     Actual Freq   Zipf Frac    Zipf Freq   Difference  Percentage  ")
    for index, item in enumerate(zipf_table, start=1):
        print("  {:3}        {:12}  {:8.0f} {:>10} {:10.2f} {:10.2f}    {:7.2f}% ".format(
            index, 
            item["word"], 
            item["actual_frequency"], 
            item["relative_frequency"], 
            item["zipf_frequency"], 
            item["difference_actual"], 
            item["difference_percent"]))
data = urllib2.urlopen("https://www.tekstowo.pl/piosenka,piwnica_pod_baranami,dezyderata.html")
soup = BeautifulSoup(data, 'html.parser')

lyrics = extractText(soup)
zipf_table = generateZipfTable(lyrics, 15)
printZipfTable(zipf_table)
print()

text = urllib2.urlopen("https://www.gutenberg.org/files/4300/4300-0.txt").read().decode("utf-8")
zipf_table = generateZipfTable(text, 15)
printZipfTable(zipf_table)
````
