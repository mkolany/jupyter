```python
from ipywidgets import *
from traitlets import *
# import traitlets as tl
import urllib.request as urllib2
from bs4 import BeautifulSoup
import validators
from tkinter import Tk, filedialog
import pandas as pd

class ControlPanelRow(HBox):
    def __init__(self, children):
        super().__init__(children=list(children))
        self.layout=box_layout = Layout(display='flex',
                    flex_flow='row',
                    align_items='stretch',
                    justify_items='center',
#                     justify_content='space-around', #'space-between',
#                     justify_content='space-between', #'space-between',
                    border='1px solid lightgray')
        
#         for w in self.children:
#             print(w._model_name)
#             w.layout.width='100%' if (w._model_name != 'ToggleButtonModel') else '80px'
#             print(w.trait_names())
    
class ControlPanelCol(VBox):
    def __init__(self, children, label = False):
        super().__init__(children=list(children))
    def addRow(self, *args, **kwargs):
        child = self.children[-1].__class__( **kwargs)
        self.children += (child, )
            
class Student(HasTraits):
    name = Unicode() 
    surname = Unicode() 
    email = Unicode() 
    selected = Bool()
    
    def __init__(self, name='', surname='', email='', selected=False):  
        self.name = name 
        self.surname = surname
        self.email = email
        self.selected = selected

    @observe('name', 'surname', 'email', 'selected')
    def _observe_name(self, change):
        print('{} changed: {}=>{}'.format(change['name'], change['old'],change['new']))
        
class StudentRow(ControlPanelRow):
    def __init__(self, name='', surname='', email='', selected=False):  
        self.s=Student(name, surname, email, selected)
        
        self.name = Text(placeholder='name', value=self.s.name)                 # description='name'
        self.surname = Text(placeholder='surname', value=self.s.surname)              # description='surname'
        self.email = Text(placeholder='email', value=self.s.email)                # description='email'
        self.selected = ToggleButton(icon='check', value=self.s.selected)   

        
        link( (self.name, "value"), (self.s, "name"))
        link(  (self.surname, "value"), (self.s, "surname"))
        link( (self.email, "value"), (self.s, "email"),)
        link(  (self.selected, "value"), (self.s, "selected"))
        
        row = [self.name, self.surname, self.email, self.selected]
        
        super().__init__(children=row)
        
        
        for w in self.children:
#             print(w._model_name)
            w.layout.width='100%' if (w._model_name != 'ToggleButtonModel') else '80px'
            
    def getLabelRow(self):
        row=[Button() for _ in range(4)]
        labelRow = ControlPanelRow(children=row)
        for k, w in enumerate(labelRow.children):
            w2 = self.children[k]
            traits = w2.trait_names()
            if 'layout' in traits:
                w.layout.width = w2.layout.width
            if 'placeholder' in traits: 
                w.value = w2.placeholder
                w.description = w2.placeholder
            else:
                w.value = ''
                
            w.layout.display = 'flex'    
            w.layout.justify_content = 'center'
            w.button_style = 'primary' #'info'
        return labelRow
    
    def getRow(self):
        return self.s.__dict__['_trait_values']
      
s=StudentRow()
l=s.getLabelRow()

cp=ControlPanelCol(children=[l,s], label = True)

b=Button(description='Add Student', button_style = 'success')
b.on_click(lambda b: cp.addRow())

def loadCSV(b,index_col=None):
    root = Tk()
    root.withdraw()                                        
    root.call('wm', 'attributes', '.', '-topmost', True)   
    files = filedialog.askopenfilename(multiple=True, title = "Select student list file",filetypes = (("CSV files","*.csv"),("TXT files","*.txt"),))  
    for file in files:
        fob=open(file,'r')
        df = pd.read_csv(fob, index_col=index_col)        
        rowsDict = df.to_dict(orient='index')
        for row in rowsDict.values():
            cp.addRow(name=row['name'], surname=row['surname'], email=row['email']) 
        
loadCSVButton=Button(description='Load CSV', button_style = 'info')
loadCSVButton.on_click(loadCSV)


d=Button(description='Save CSV', button_style = 'info')
# d.on_click(lambda b: print())

cr = ControlPanelRow([b, loadCSVButton, d])

ControlPanelCol(children=[cr,cp])
```
