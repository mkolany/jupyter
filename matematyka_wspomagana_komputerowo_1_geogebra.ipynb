## Oswajanie się z geogebrą

### Wstęp teoretyczny 
#### Geometria euklidesowa przedstawiana jest jako system aksjomatyczny, w którym wszystkie twierdzenia muszą wynikać z aksjomatów, czyli zdań przyjmowanych z góry jako prawdziwe.

> W podanym przez siebie systemie Euklides wyróżnił pięć aksjomatów lub pewników płaszczyzny nazywanej później również euklidesową:
> 
> 1. Dowolne dwa punkty można połączyć odcinkiem.
> 
> 2. Dowolny odcinek można przedłużyć nieograniczenie (uzyskując prostą).
> 
> 3. Dla danego odcinka można zaznaczyć okrąg o środku w jednym z jego końcowych punktów i promieniu równym jego długości.
> 
> 4. Wszystkie kąty proste są przystające.
> 
> 5. Dwie proste, które przecinają trzecią w taki sposób, że suma kątów wewnętrznych po jednej stronie jest mniejsza od dwóch kątów prostych, przetną się z tej właśnie strony.

### Uwaga 
#### Dla geometrii na płaszczyźnie piąty z aksjomatów, tzw. postulat Euklidesa lub postulat równoległości, można sformułować również następująco:
> 
> „przez dany punkt nienależący do danej prostej można poprowadzić jedną prostą rozłączną z daną prostą”
>

# Podstawowe zagadnienia związane z używaniem geogebry:

1. Można to zrobić ręcznie. W tym celu należy w zasobniku (toolbarze) znaleć narzędzie "Point". Nowy punkt wstawiamy klikając w obszarze roboczym

Np.:
* punkt o współrzędnych (1,2) można uzyskać w następujący sposób:
     
     `Point({1,2})` 

       
     `(1,2)`  

     `A=(1,2)`   


###  A co jeśli nie chcemy podawać konkretnych współrzędnych?
* losowy punkt należący do obszaru $(x_{Min},x_{Max}) \times (y_{Min},y_{Max})$:

     `RandomPointIn( <xMin>, <xMax>, <yMin>, <yMax> )`    



     `(RandomUniform(-5,5),RandomUniform(-5,5))`

     `Point(Sequence(RandomUniform(-5,5),n,1,2))`

* lista dziesięciu losowych punktów z obszaru $(-5,5)\times(-5,5)$:

     `l=Zip(RandomPointIn(-5,5,-5,5),i,Sequence(10))`  


* losowy punkt z wnętrza wielokąta, o wierzchołkach będących elementami listy, którą otrzymaliśmy w poprzednim kroku (BTW. wielokąt ten można zobaczyć pisząc `Polygon(l)`):

     `RandomPointIn(l)` 


###  A co jeśli nie chcemy, żeby rezultat był aż taki losowy?
* jakiś punkt należący do okręgu o środku w punkcie $(0,0)$ i promieniu $1$:

     `A=Point(Circle(O,1))`    


     1. `c=Circle(O,1)`
     2. `Point(c)`

*  punkty postaci $(n,-n)$, gdzie $n\in(-5,5)\cap\mathbb{Z}$:

     `l1=Sequence((n,-n),n,-5,5)`


     lub (druga propozycja jest dość buńczuczna):      
     `Zip((n,-n),n,(Sequence(-5,5)))`
     `Zip((n,m),n,Sequence(-5,5),m,Sequence(5,-5))`  
     `Point(RandomUniform(-5,5,2))`


* wszystkie punkty kratowe z obszaru $(-5,5)\times(-5,5)$

     `l2=Sequence(Sequence((n,m),n,-5,5),m,-5,5)`    


     lub:
     `Zip(Zip((n,m),n,Sequence(-5,5)),m,Sequence(-5,5))`
     
     warto zaznaczyć, że dzięki potędze matematyki możemy uwolnić się od zagnieżdżania pętli w innej pętli:    
     >   `Zip((floor(((n)/(11)))-5,Mod(n,11)-5),n,Sequence(0,11*11-1))`  
    


###  Czemu te komendy wyglądają aż tak paskudnie?
Przepraszam... Tak czy owak, wyklikanie tego myszką, jest czasem jeszcze mniej wygodne. Spróbuj odtworzyć ręcznie wynik następujących komend:
* `Polygon(Sequence(Point(Circle((0,0),n),fractionalPart(k n)),n,1,25,j))`
* `Reflect(Polygon(Sequence(Point(Circle((0,0),n),fractionalPart(k n)),n,1,25,j)),Circle(O,5))`

np. dla $k=1\frac15,\; j=\frac12$. Co to w ogóle jest?

###  Jak wstawiać odcinki i linie?
* `Segment(A,B)`  
     `Segment((0,0),(1,1))`
* `Line(A,B)`  
     `Line((0,0),(1,1))`

###  Jak zrobić prostą równoległą do innej prostej przechodzącą przez punkt? A prostopadłą?
*  prosta równoległa do $l$ przechodząca przez $A$

      `Line(A,l)`  

*  prosta prostopadła (...):

     `PerpendicularLine(A,l)`
