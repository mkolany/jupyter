```python
%%manim CircularInversion1

def circInv2d(pos,cx=0,cy=0, cr=1):
    x,y,z=pos
    dx = x - cx; dy = y - cy
    d = np.hypot(dx, dy)
    
    if (d!=0):
        dx = dx / d; dy = dy / d
        dnew = cr ** 2.0 / d
        x = float(dnew * dx + cx)
        y = float(dnew * dy + cy)

        return [x,y,z]
    else:
        return [0,0,0]
        
class CircularInversion1(MovingCameraScene):
    def construct(self):
        func =  lambda z: ORIGIN-np.sin(z[0]) * UR + np.cos(z[1]) * LEFT + z / 10
        vector_field = ArrowVectorField(
                func, min_color_scheme_value=2, max_color_scheme_value=10, colors=[WHITE, BLACK]
            )
        c=Circle(radius=1, color=WHITE)
        mob = ImageMobject('papaj.png').scale(0)
    
        vg=VGroup(*[Circle(color=WHITE) for _ in range(10**3 +1)])
        vg2=VGroup(*[TangentLine(c, alpha=angle, length=30) for angle in np.linspace(0,1,num=36, endpoint=False)])
        vg.arrange_in_grid(buff=(0,0))
        vg.move_to(ORIGIN).shift(LEFT+UP)
        poly_hex = RegularPolygon(n=6, start_angle=30*DEGREES).scale(3).shift(3*LEFT)
        
#         self.play(*[Create(m) for m in [ vg ]])
        self.add(c,vg,vg2)
        self.play(
            vg.animate.apply_function(
                lambda p: circInv2d(p)
            ),
#             FadeIn(mob)
        )
#         mob.add_updater(vector_field.get_nudge_updater(pointwise=True))
        self.add(mob)
        self.play(ApplyMethod(self.camera.frame.scale, 16), FadeOut(c,vg,vg2), ApplyMethod(mob.scale, 16))
    
        self.wait()
        self.play(FadeOut(mob))
    
```

```python
%%manim Polygrow

class Polygrow(Scene):
    def construct(self):
        poly_1 = RegularPolygon(n=6)
        poly_2 = RegularPolygon(n=6, start_angle=30*DEGREES, color=GREEN)
        poly_3 = RegularPolygon(n=10, color=RED)

        poly_group = Group(poly_1, poly_2, poly_3).scale(1.5).arrange(buff=1)
        self.add(poly_group)

        pentagram = RegularPolygram(5, radius=2)
        star = Star(outer_radius=2, color=RED)

        self.add(pentagram)
        self.play(Create(star), run_time=3)
        self.play(FadeOut(star), run_time=2)
```

```python
%%manim Coloring


class Coloring(Scene):
    def construct(self):
        func = lambda pos: pos - LEFT * 5
        colors = [RED, YELLOW, BLUE, DARK_GRAY]
        min_radius = Circle(radius=2, color=colors[0]).shift(LEFT * 5)
        max_radius = Circle(radius=10, color=colors[-1]).shift(LEFT * 5)
        vf = ArrowVectorField(
            func, min_color_scheme_value=25, max_color_scheme_value=40, colors=palette
        )
        self.add(vf, min_radius, max_radius)
```

```python
%%manim ColoringpawningAndFlowingArea

class ColoringpawningAndFlowingArea(Scene):
    def construct(self):
        func = lambda pos: np.sin(pos[0]) * UR + np.cos(pos[1]) * LEFT + pos / 5
        stream_lines = StreamLines(
            func, x_range=[-3, 3, 0.2], y_range=[-2, 2, 0.2], padding=1
        )
        poly_hex = RegularPolygon(n=6, start_angle=30*DEGREES).scale(3)

        spawning_area = Rectangle(width=6, height=4)
        flowing_area = Rectangle(width=8, height=6)
        labels = [Tex("Spawning Area"), Tex("Flowing Area").shift(DOWN * 2.5)]
        for lbl in labels:
            lbl.add_background_rectangle(opacity=0.6, buff=0.05)

        self.add(stream_lines, poly_hex, spawning_area, flowing_area, *labels)
```
